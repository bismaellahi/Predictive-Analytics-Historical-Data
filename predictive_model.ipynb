```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load dataset
data = pd.read_csv("sales_data.csv")

# Show first 5 rows
print(data.head())
```

```python
# Features and Target
X = data[['Month']]
y = data['Sales']

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Predict Existing Data
predictions = model.predict(X)

print(predictions)
```

```python
# Predict next 6 months

future_months = pd.DataFrame({
    'Month': [13, 14, 15, 16, 17, 18]
})

future_predictions = model.predict(future_months)

future = pd.DataFrame({
    'Month': future_months['Month'],
    'Predicted Sales': future_predictions
})

print(future)
```

```python
# Plot actual and predicted values

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(data['Month'], data['Sales'], color='blue', label='Actual Sales')
plt.plot(data['Month'], predictions, color='red', label='Regression Line')

plt.scatter(
    future_months['Month'],
    future_predictions,
    color='green',
    label='Future Prediction'
)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Forecast Using Linear Regression")

plt.legend()

plt.show()
```

```python
# Check model accuracy

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(data['Month'], data['Sales'], label='Actual Sales')
plt.plot(data['Month'], predictions, label='Regression Line')
plt.scatter(future_months['Month'], future_predictions, label='Future Prediction')

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Forecast Using Linear Regression")
plt.legend()

plt.savefig("output.png")
plt.show()
```
